#  A SMART DOOR ACCESS SYSTEM FOR UMaT ADMIN STAFF

```python
class Staff:
    def __init__(self,name,access_code):
        self.staff_name = name
        self.__access_code = access_code

    @property
    def access_code(self):
        return self.__access_code

    @access_code.setter
    def access_code(self,new_access_code):
        self.__access_code = new_access_code

    def display_info(self):
        print(f"Hello {self.staff_name} your access code is {self.access_code}")

staff1=Staff("Matthew Cobinnah","CE154")
staff1.access_code="FOE21"
staff1.display_info()
```
